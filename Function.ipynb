**Q.1. What is the difference between a function and a method in Python?**

1. Function

A function is a block of code that performs a specific task.
It is defined independently (outside of a class) using def keyword.

Example of a Function:

def add(a,b):
  return a+b
result = add(5,3)
print(result)   # Output: 8

Here,

add() is a function

It does not belong to any class.

You call it directly using its name.

2. Method

A method is a function that belongs to a class.
It always works with an object of that class.

Example of a Method:

class Calculator:
  def add(self,a,b):  #this is method
    return a+b
calc = calculator()  #creating object
result = calc.add(5,3)  #calling method using object
print(result)  #output 8

Here,

add() is a method because it is defined inside a class.

It takes self as its first parameter (refers to the object).

We call it using an object → calc.add().

| Feature         | Function                 | Method                           |
| --------------- | ------------------------ | -------------------------------- |
| Defined         | Outside a class          | Inside a class                   |
| Calls           | Directly (e.g., `add()`) | Using object (e.g., `obj.add()`) |
| First parameter | No special parameter     | Uses `self`                      |
| Belongs to      | No object                | Belongs to an object or class    |

**Q4. What is the purpose of the `return` statement in a Python function.**

The purpose of the return statement in a Python function is to send a value back to the place where the function was called and end the function's execution.

**Why do we use return?**
1. To give output from the function

A function can perform calculations and send the result back using return.

2. To stop the function execution

Anything written after return will not run.

3. To use the function’s result somewhere else

You can store the returned value in a variable or use it in another calculation.

**Example 1: Function with return**

def add(a, b):

    return a + b

result = add(5, 3)

print(result)   # Output: 8

Here, return a + b sends 8 back to the caller.

**Example 2: return stops the function**

def test():

The print inside the function will not run after return.

    print(a + b)

print(result)

8

**Q.5. What are iterators in Python and how do they differ from iterables?**

In Python, iterators and iterables are related but distinct concepts.

- An iterable is an object that can be looped over, such as lists, tuples, or strings. It implements the __iter__() method, which returns an iterator.
- An iterator is an object that actually fetches items one at a time from an iterable.

An iterator must have:

__iter__() → returns the iterator object itself

__next__() → returns the next value

When there are no values left, __next__() raises StopIteration.

Key differences:

- Iterables can be iterated over multiple times, while iterators can only be iterated over once.
- Iterators maintain their state, whereas iterables do not.

# Iterable (list)

my_list = [1, 2, 3]

for item in my_list:

    print(item)  # Can iterate multiple times

# Iterator

my_iter = iter(my_list)

print(next(my_iter))  # Output: 1

print(next(my_iter))  # Output: 2

print(next(my_iter))  # Output: 3

next(my_iter)  # Raises StopIteration exception

**Q.6. Explain the concept of generators in Python and how they are defined.**

- In Python, generators are a type of iterable, similar to lists or tuples. However, unlike lists, generators don't store all values in memory at once. Instead, they generate values on-the-fly, making them memory-efficient.

- Generators are defined using a function with the yield keyword instead of return. When a generator is called, it returns an iterator that produces the yielded values.


def infinite_sequence():

    num = 0

    while True:

        yield num

        num += 1

gen = infinite_sequence()

print(next(gen))  # Output: 0

print(next(gen))  # Output: 1

print(next(gen))  # Output: 2


Benefits of generators:

- Memory efficiency: Generators use constant memory, making them suitable for large datasets.
- Lazy evaluation: Values are generated only when needed, reducing computation.

You can also define generators using generator expressions, similar to list comprehensions:


gen_expr = (x**2 for x in range(10))

print(next(gen_expr))  # Output: 0

print(next(gen_expr))  # Output: 1


**Q.7. What are the advantages of using generators over regular functions?**

Generators in Python provide several advantages over regular functions, especially when dealing with large data, infinite sequences, or performance-critical applications.

Here are the key advantages:

**1. Memory Efficiency (No need to store entire data in memory)**

A normal function that returns a list creates and stores all values in memory.
A generator produces one value at a time, only when needed.

Example:

def normal_func():

    return [i for i in range(1_000_000)]

def generator_func():

    for i in range(1_000_000):

        yield i


normal_func() → uses a lot of memory

generator_func() → uses almost no extra memory

Perfect for large datasets or streams of data.

**2. Faster Execution for Large Data (Lazy evaluation)**

Generators compute values on demand (lazy evaluation), so they can start producing results immediately without waiting to generate the whole list.

**3. Can Produce Infinite Sequences**

Regular functions cannot return infinite lists — they will run forever or crash.
Generators can generate infinite sequences safely.

Example:

def infinite_numbers():

    n = 1
    while True:
        yield n
        n += 1

**4. Improves Code Readability**

Generators allow using a function-like syntax (yield) to write compact, clear iterators without writing __iter__() and __next__() manually.

**5. Better Performance in Pipelines**

Generators can be chained to create efficient data pipelines.

Example:

numbers = (i for i in range(100000))

evens = (n for n in numbers if n % 2 == 0)


This avoids creating intermediate lists → saves memory and time.

**Q.8. What is a lambda function in Python and when is it typically used?**

A lambda function in Python is a small, anonymous function defined with the lambda keyword. It's used for short, one-time use cases.


**Regular function**

def add(x, y):

**Q.9. Explain the purpose and usage of the `map()` function in Python.**

The map() function in Python applies a given function to each item of an iterable (like a list or tuple) and returns a map object.


def square(x):

    return x**2

numbers = [1, 2, 3, 4, 5]

squares = map(square, numbers)

print(list(squares))  # Output: [1, 4, 9, 16, 25]


You can also use lambda functions with map():


numbers = [1, 2, 3, 4, 5]

squares = map(lambda x: x**2, numbers)

print(list(squares))  # Output: [1, 4, 9, 16, 25]


map() is often used for:

- Data transformation: Applying a function to each item in a dataset.
- Type conversion: Converting data types, like strings to integers.
- Simplified code: Replacing loops with concise, readable code.

Note: In Python 3, map() returns a map object, which is an iterator. To get a list, you need to explicitly convert it using list().

**Q.10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

In Python, map(), reduce(), and filter() are three fundamental functions used for functional programming.

- map(function, iterable): Applies a function to each item in an iterable and returns a map object with the results.
- filter(function, iterable): Applies a function to each item in an iterable and returns a filter object with items for which the function returns True.
- reduce(function, iterable): Applies a function to the first two items in an iterable, then to the result and the next item, and so on, returning a single value.


from functools import reduce

numbers = [1, 2, 3, 4, 5]

**map()**

squares = list(map(lambda x: x**2, numbers))

print(squares)  # Output: [1, 4, 9, 16, 25]

**filter()**

evens = list(filter(lambda x: x % 2 == 0, numbers))

print(evens)  # Output: [2, 4]

**reduce()**

product = reduce(lambda x, y: x * y, numbers)

print(product)  # Output: 120


Use map() for transformations, filter() for selections, and reduce() for aggregations.

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
#the list.

def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:   # Check if the number is even
            total += num
    return total

# Example usage:
my_list = [10, 15, 22, 33, 40]
result = sum_of_even_numbers(my_list)
print("Sum of even numbers:", result)

```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

#Function to Check if a Number Is Prime

def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

#Check and Print Prime Status for Numbers 1 to 200

for n in range(1, 201):
    print(n, "=>", "Prime" if is_prime(n) else "Not Prime")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.

#Fibonacci Iterator Class

class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms   # total terms to generate
        self.count = 0           # counter
        self.a = 0               # first Fibonacci number
        self.b = 1               # second Fibonacci number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_num = self.a + self.b
            self.a, self.b = self.b, next_num
            self.count += 1
            return next_num

#Example Usage

fib = Fibonacci(10)  # generate first 10 Fibonacci numbers

for num in fib:
    print(num, end=" ")

```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

#Generator Function for Powers of 2

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

#Example Usage

for value in powers_of_two(5):
    print(value, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

#Generator Function to Read a File Line by Line

def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.rstrip("\n")   # yields each line without the trailing newline

#Example Usage

for line in read_file_lines("sample.txt"):
    print(line)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius = [0, 10, 25, 37, 100]

# Convert to Fahrenheit using map()
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

text = "Python Programming"

# Define vowels
vowels = "aeiouAEIOU"

# Use filter to remove vowels
result = "".join(filter(lambda ch: ch not in vowels, text))

print("Original:", text)
print("Without vowels:", result)
```
